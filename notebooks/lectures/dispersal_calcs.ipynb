```python
from sympy import *
```

```python
S,d,n,c,n_m,d_m=symbols('S,d,n,c,n_m,d_m')
prr=(S*(1-d) + (n-1)*d*(1-c))/(S*(1-d) + (n-1)*d*(1-c) + n_m*d_m*(1-c))
pmm=(S*(1-d_m) + (n_m-1)*d_m*(1-c))/(S*(1-d_m) + (n_m-1)*d_m*(1-c) + n*d*(1-c))
pre=n*d/(n*d + n_m*d_m)
```

```python
nnext=n*prr + n_m*(1 - pmm) + (S - n - n_m)*pre
```

```python
diff(nnext,n).subs(n_m,0)
```

```python
nmnext = n*(1-prr) + n_m*pmm + (S - n - n_m)*(1-pre)
```

```python
inv = diff(nmnext,n_m).subs(n_m,0).subs(n,S)
inv
```

```python
(inv-1).factor()
```

```python
s = symbols('s')
invapp = limit(inv.subs(S,1/s),s,0)
```

```python
(invapp-1).factor()
```

```python
diff(inv,d_m).subs(d_m,d).factor()
```
