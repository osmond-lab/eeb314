## Model

We'll keep analyzing the same equations

$\displaystyle{n(t+1) = n(t) \left(1 + r \left(1 - \frac{n(t) + \alpha(s,s_m) n_m(t)}{K(s)}\right)\right)}$

$\displaystyle{n_m(t+1) = n_m(t) \left(1 + r \left(1 - \frac{n_m(t) + \alpha(s_m,s) n(t)}{K(s_m)}\right)\right)}$,

But now choose some functions for carrying capacity $K$ and competition $\alpha$.

Let's assume carrying capacity is a Gaussian ("bell-shaped") function,

$\displaystyle{K(s) = K_0 e^{-(s-s_0)^2 / a_K}}$

where $s_0$ is the size of the most abundant algae, $K_0$ is the carrying capacity for daphnia that best consume the most abundant algae, and $a_K$ describes the variation in algae size (larger $a_K$ means a larger range of available sizes).

```python
from sympy import *

K0,s0,aK,s = var('K_0,s_0,a_K,s')

def K(s):
  'carrying capacity as a function of algae size'
  return K0*exp(-(s-s0)**2/aK)

Ks = K(s).subs(K0,50).subs(s0,10)
p = plot(Ks.subs(aK,10),Ks.subs(aK,1),(s,0,20),xlabel='$s$',ylabel='$K(s)$',legend=True,label=['$a_K=10$','$a_K=1$'], show=False)
p.show()
```

Let's assume competition coefficients are also Gaussian,

$\displaystyle{\alpha(s_i,s_j) = e^{-(s_i-s_j)^2/a_\alpha}}$.

This equation implies that individuals compete more strongly with individuals that have more similar trait values and they compete more broadly when $a_\alpha$ is larger (i.e., smaller $a_\alpha$ means they are more specialized).

```python
aa = var('a_a')

def a(sm,s):
  'competition coefficient as a function of the trait values of the competitors'
  return exp(-(sm-s)**2/aa)

p = plot(a(10,s).subs(aa,10),a(10,s).subs(aa,1),(s,0,20),xlabel='$s$',ylabel=r'$\alpha(s_m,s)$',legend=True,label=[r'$a_\alpha=10$',r'$a_\alpha=1$'], show=False)
p.show()
```

## Bonus

We can visualize both evolutionary and convergence stability with Pairwise Invasibility Plots (PIP's). These plots have resident trait values on the $x$ and and mutant trait values on the $y$. We color in the mutant trait values that can invade a given resident trait value (in this case, in blue).

```python
import numpy as np
from matplotlib import pyplot as plt

# invasion fitness
def inv_fun(s_m,s,r=0.75,a_a=10,a_K=5,s_0=10):
  invasionfitness = -r*exp(-s**2/a_a + 2*s*s_m/a_a - s_m**2/a_a - s**2/a_K + 2*s*s_0/a_K - 2*s_0*s_m/a_K + s_m**2/a_K) + r + 1
  if invasionfitness>1:
    return 1 # return 1 if mutant invades
  else:
    return 0 # return 0 if mutant does not invade

# evaluate
def compute_pip(r=0.75,a_a=10,a_K=5,s_0=10,smin=5,smax=15,steps=100):
    ss = np.linspace(smin,smax,steps)
    X,Y = np.meshgrid(ss,ss) # X and Y values
    # store the invasion success in a matrix
    PIP = []
    for sm in ss:
      row = []
      for s in ss:
        row.append(inv_fun(sm,s,r,a_a,a_K,s_0)) # invasion success of sm into s
      PIP.append(row)
    return X,Y,PIP

# plot
def plotfun(X,Y,Z):
  fig, ax=plt.subplots(1,1,figsize=(5,5))
  ax.set_xlabel('$s$')
  ax.set_ylabel('$s_m$')
  ax.contourf(X,Y,Z, colors=['white','black','blue','black'])
  ax.plot(X[0],X[0],'k',lw=5)
```

```python
a_a=10
a_K=5
print('a_a=%s, a_K=%s' %(a_a,a_K))
X,Y,Z=compute_pip(a_a=a_a,a_K=a_K)
plotfun(X,Y,Z)
```

```python
a_a=5
a_K=10
print('a_a=%s, a_K=%s' %(a_a,a_K))
X,Y,Z=compute_pip(a_a=a_a,a_K=a_K)
plotfun(X,Y,Z)
```

From these plots we can read off how evolution will proceed by small mutational steps by looking along the 1:1 line (thick black). In particular, say we start from a small resident trait value, eg, $s=6$. We see that in either case mutants with trait values larger than 6 can invade because the PIP is blue above the 1:1 line at $s=6$. That means the trait value will increase and we shift to the right. We keep moving right until the PIP directly above the 1:1 line turns white. We similarly move left when the PIP is blue below the 1:1 line, eg, at $s=14$. This process leads us to convergence stable singular strategies. Now, once we're at a convergence stable strategy, evolutionary stability is determined by the color of the PIP directly above and below. When the PIP is white directly above and below the singular strategy it cannot be invaded by mutants in either direction, so it is evolutionarily stable. In contrast, when the PIP is blue directly above and below the singular strategy it can be invaded by mutants on either side of it, leading to evolutionary branching.
